```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    list_of_genres = []
    for i in ast.literal_eval(obj):
        list_of_genres.append(i['name'])
    return list_of_genres
```

```python
data['Genres'] = data['genres'].apply(convert)
```

```python
data['Keywords'] = data['keywords'].apply(convert)
```

```python
def convertCast(obj):
    actors = []
    for i in ast.literal_eval(obj):
        actors.append(i['name'])
        actors = actors[0:3]
    return actors
```

```python
data['Top3Cast'] = data['cast'].apply(convertCast)
```

```python
crew = ast.literal_eval(data['crew'][0])
```

```python
def convertCrew(obj):
    director = []
    crew = ast.literal_eval(obj)
    for i in crew:
        if i['job']=='Director':
            director.append(i['name'])
    return director
```

```python
data['Director'] = data['crew'].apply(convertCrew)
```

```python
data.drop(['genres','keywords','cast','crew'],axis=1,inplace=True)
```

```python
data['overview'] = data['overview'].apply(lambda x: x.split())
```

```python
def removeSpaces(obj):
    L = []
    for i in obj:
        L.append(i.replace(" ",""))
    return L
```

```python
data['Top3Cast'] = data['Top3Cast'].apply(removeSpaces)
data['Genres'] = data['Genres'].apply(removeSpaces)
data['Keywords'] = data['Keywords'].apply(removeSpaces)
data['Director'] = data['Director'].apply(removeSpaces)
```

```python
#Now we will create a tag column
data['tags'] = data['overview'] + data['Genres'] +data['Keywords'] + data['Top3Cast'] + data['Director']
```

```python
df = data[['id','original_title','tags']]
```

```python
#Converting to string
df.loc[:, 'tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
#Converting to lowercase
df.loc[:, 'tags'] = df['tags'].apply(lambda x: x.lower())
```

**Stemming: technique of removing inflection from the words and bringing all words to their root.**
**Types of stemming:**
- Porter Stemmer
- Lancaster Stemmer
- Snowball Stemmer

```python
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
from nltk.stem import SnowballStemmer
```

```python
#Most primitive method of stemming
ps = PorterStemmer()
words = ['eating','eats','eaten','puts','putting']
for word in words:
    print(word,'--->',ps.stem(word))
```

```python
#Can lead to over stemming sometimes
ls = LancasterStemmer()
words = ['eating','eats','eaten','puts','putting']
for word in words:
    print(word,'--->',ls.stem(word))
```

```python
#Can lead to over stemming sometimes
ss = SnowballStemmer(language='english')
words = ['eating','eats','eaten','puts','putting']
for word in words:
    print(word,'--->',ss.stem(word))
```

```python
#Now let's implement it here on our data.
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
sorted(cv.vocabulary_)
```

**Steps involved in recommendation**
1. Stemming
2. CountVectorizer()
3. Calculate distance using cosine_similarity
4. Define a recommend function which gives you the first 5 movies that are similar to a movie A.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = df[df['original_title']== movie].index[0]
    movie_distances = similarity[movie_index]
    recommended_list = sorted(list(enumerate(movie_distances)),reverse=True,key=lambda x:x[1])[1:6]
    for movies in recommended_list:
        print(df.iloc[movies[0]]['original_title'])
```

```python
recommend('Avatar')
```

```python
recommend('The Dark Knight Rises')
```
